```python
import requests
from bs4 import BeautifulSoup
import re
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# 設定中文字型
font_path = r"C:\Users\jessi\OneDrive\GitHub\XHR\NotoSansTC-VariableFont_wght.ttf"
font_prop = FontProperties(fname=font_path, size=12)

# 王凱的維基百科頁面 URL
wiki_url = "https://zh.wikipedia.org/zh-tw/%E7%8E%8B%E5%87%B1_(%E4%B8%AD%E5%9C%8B%E5%A4%A7%E9%99%B8%E6%BC%94%E5%93%A1)"

# 發送 HTTP 請求並取得頁面內容
response = requests.get(wiki_url)
html_content = response.content

# 使用 BeautifulSoup 解析頁面内容
soup = BeautifulSoup(response.text, 'html.parser')

# 找到簡介部分
intro_section = soup.find('div', {'class': 'mw-parser-output'})

# 正規表達式模式，匹配年份及後面的一句文字
pattern = re.compile(r'(\d{4}年.*?。)')

# 在簡介文本中搜索匹配的內容
matches = pattern.findall(intro_section.text)

# 列印匹配的結果，排除指定句子
filtered_matches = [match for match in matches if "1982年8月18日" not in match]

# 按照年份排序
sorted_matches = sorted(filtered_matches, key=lambda x: int(re.search(r'\d{4}', x).group()))

# 為每個事件分配編號，從A到Z
events = [{"id": chr(idx), "time": re.search(r'\d{4}年', match).group()} for idx, match in enumerate(sorted_matches, start=65)]
for idx, match in enumerate(sorted_matches, start=65):  # 使用ASCII碼中的大寫字母A-Z
    print(f"{chr(idx)}: {match}")
    
# 繪製時間軸
fig, ax = plt.subplots(figsize=(10, 2))
event_positions = {}  # 用來儲存每年事件的垂直位置

for event in events:
    year = event["time"][:-1]
    if year not in event_positions:
        event_positions[year] = 0
    else:
        event_positions[year] -= 0.1  # 處理同一年的事件位置

    ax.plot(year, event_positions[year], 's', color='gold')  # 在時間軸上繪製事件點
    ax.text(year, event_positions[year], f' {event["id"]}', ha='center', va='center', fontproperties=FontProperties(fname=font_path, size=16, weight='bold'))  # 顯示事件編號，字母大小為16

# 移除上面及左右的邊框
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

# 設定圖表樣式
plt.yticks([])  # 不顯示y軸
plt.title('王凱事業時間軸', fontproperties=font_prop)
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 王凱的維基百科頁面 URL
wiki_url = "https://zh.wikipedia.org/zh-tw/%E7%8E%8B%E5%87%B1_(%E4%B8%AD%E5%9C%8B%E5%A4%A7%E9%99%B8%E6%BC%94%E5%93%A1)"
response = requests.get(wiki_url)
soup = BeautifulSoup(response.text, 'html.parser')

# 找到所有具有 "wikitable" 類別的表格
wikitables = soup.find_all('table', class_='wikitable')

# 如果有至少兩個具有 "wikitable" 類別的表格
if len(wikitables) >= 2:
    # 選取第二個表格
    second_table = wikitables[2]

    # 读取表格数据
    df = pd.read_html(str(second_table))[0]
  
    # 删除指定列
    columns_to_drop = ["年份","播出時間", "備註"]
    df = df.drop(columns=columns_to_drop, axis=3)
    
    # 將 DataFrame 轉為文本形式
    table_text = df.to_string(index=False)

    # 列印處理後的文本内容
    print(table_text)
   
```

```python
import re

# 要刪除的字詞列表
words_to_remove = ["節　目", "-爪哇文化巡禮","-瘋狂在日惹","-天堂峇里島","-饕餮峇里島","-去，你的印尼海島","-原味冒險島","第四季～王凱慢游芬蘭","第四期","二", "第三季","NaN"]

# 保留《》內容的條件
def should_keep_inside_brackets(content):
    # 這裡可以根據需要定義條件，例如保留包含特定文字的內容
    return "中國文藝志願者在行動", "朗讀者" in content

# 定義一個函數，用於處理每個節目名稱
def process_program_name(name):
    # 使用正則表達式找到並刪除《》中的文字
    name = re.sub(r'《.*?》', '', name)
    
    for word in words_to_remove:
        name = name.replace(word, '')
    return name.strip()

# 對每一行進行處理
lines = table_text.split('\n')
processed_lines = [process_program_name(line) for line in lines]

# 去除空行
processed_lines = [line for line in processed_lines if line]

# 找到最長的節目名稱，以便靠左對齊
max_len = max(len(line) for line in processed_lines)

# 靠左對齊顯示處理後的結果
for line in processed_lines:
    print(f"{line.ljust(max_len)}")
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from collections import Counter

# 設定中文字型
font_path = r"C:\Users\jessi\OneDrive\GitHub\XHR\NotoSansTC-VariableFont_wght.ttf"
font = FontProperties(fname=font_path, size=48)

# 進行字詞統計
word_counts = Counter(processed_lines)

# 列印字詞統計結果
print("\n字詞統計：")
for word, count in word_counts.items():
    print(f"{word}: {count} 次")

# 生成長條圖
plt.figure(figsize=(48, 24))  # 調整圖表大小
plt.bar(word_counts.keys(), word_counts.values(), color='gold')
plt.xlabel('電視台', fontproperties=font)  # 使用中文字型
plt.ylabel('出現次數', fontproperties=font)  # 使用中文字型
plt.title('電視台出現次數統計', fontproperties=font)  # 使用中文字型
plt.xticks(rotation=-45, ha='left', fontproperties=font)  # 調整 x 軸標籤的角度，使用中文字型

# 調整 y 軸刻度數字字體大小
plt.tick_params(axis='y', labelsize=48)


plt.tight_layout()  # 自動調整布局，避免標籤重疊
plt.show()
```
